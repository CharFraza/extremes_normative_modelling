This script is created to run several normative models in a parallel manner.

```python
import os
import sys
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
sys.path.append('/home/preclineu/chafra/Desktop/PCNtoolkit/') # Updated
from pcntoolkit.util.utils import create_bspline_basis
sns.set(style='ticks', palette='Set2')
sns.despine()

sys.path.append('/project_freenas/3022017.02/UKB/scripts')
from ukb_utils import lookup_UKB
```

# Load current datapaths

```python
ukb_dir = '/project_freenas/3022017.02/UKB/'
phen_dir = os.path.join(ukb_dir, 'phenotypes', 'current')
data_dir = '/project_cephfs//3022017.02/projects/big_data/data/'
ukb_proc_dir = os.path.join(data_dir, 'ukb_processed_4')
os.makedirs(ukb_proc_dir, exist_ok=True)

print('reading data files...')
```

```python
IDPs = pd.read_csv(os.path.join(phen_dir, '31_brain_IDPs.csv'))#, low_memory = False) 
IDPs.head()
```

## Get demographic data age, gender and site

```python
demographics = pd.read_csv(
    '/project_freenas/3022017.02/UKB/phenotypes/current/99_miscellaneous.csv',
    usecols=['eid', '54-2.0', '21003-2.0'])
demographics['31-0.0'] = pd.read_csv(
    '/project_freenas/3022017.02/UKB/phenotypes/funpack_nov_2021_ukb49421/01_basic_demographics.csv',
    usecols=['31-0.0'])
demographics = demographics[['eid', '31-0.0', '54-2.0', '21003-2.0']]

sns.histplot(data=demographics, x="21003-2.0")
plt.title('Age participants IDPs')
print('processsing...')
```

## Drop NAN vales
The IDPs contain a lot of NAN values, so the number of subjects decreases significantly when we remove those.

```python
IDPs = IDPs.dropna(how='all', subset=IDPs.columns[1:])
X = demographics[demographics['eid'].isin(IDPs['eid'])]
eid = IDPs['eid']

# Select only instance 2 (first imaging visit)
IDPs = IDPs.filter(regex='2.0$', axis=1)
IDPs['eid'] = eid
IDPs.head()
```

## recode site column to dummy coding

```python
sites = X['54-2.0'].unique()
new_sites = pd.get_dummies(X['54-2.0'])
X = pd.concat([X,new_sites], axis=1)
X = X.drop(['54-2.0'], axis=1)
X.head()
```

```python
N = X.shape[0]
os.makedirs(ukb_proc_dir, exist_ok=True)
IDPs.to_pickle(os.path.join(ukb_proc_dir, 'IDPs_processed.pkl'))
X.to_pickle(os.path.join(ukb_proc_dir, 'cov_processed.pkl'))

idp_ids = IDPs.columns[1:].to_list()
```

```python
with open(os.path.join(ukb_proc_dir, 'idp_ids.txt'), 'w') as f:
    for idp in idp_ids:
        f.write(idp + '\n')
```

# Select train/test split 0.9 for 10/90
Here its changed to a 10-90 split, as we need more testing data for the EVT algorithm later.
Recommended setting is 0.5 or 50/50.

```python
train = np.random.uniform(size=N) > 0.90
test = np.logical_not(train)
np.savetxt(os.path.join(ukb_proc_dir, 'train_idx.txt'), train)
np.savetxt(os.path.join(ukb_proc_dir, 'test_idx.txt'), test)
print(f'Amount of training subjects: {sum(train)}')
print(f'Amount of testing subjects: {sum(test)}')
```

# Loop over all files and save the train/response data

```python
for idp in idp_ids:
    y = IDPs[['eid', idp]]

    # First get indices for non-NaNs
    nz = np.logical_not(pd.isna(y[idp]))
    tr = train[nz]
    te = test[nz]

    # remove NaNs
    y = y.dropna()
    y = y[nz]
    x = X[X['eid'].isin(y['eid'])]

    # remove outliers (abs(Z) > 7)
    yz = (y[idp] - np.mean(y[idp])) / np.std(y[idp])
    y_valid = np.abs(yz.to_numpy()) < 7

    tr = tr & y_valid
    te = te & y_valid

    idp_dir = os.path.join(ukb_proc_dir, idp)
    os.makedirs(idp_dir, exist_ok=True)

    y[tr].to_csv(os.path.join(idp_dir, 'resp_tr.txt'),
                 index=False,
                 columns=[idp],
                 sep=" ",
                 header=False)
    y[te].to_csv(os.path.join(idp_dir, 'resp_te.txt'),
                 index=False,
                 columns=[idp],
                 sep=" ",
                 header=False)
    y[tr].to_csv(os.path.join(idp_dir, 'resp_tr_index.txt'),
                 index=True,
                 sep=" ",
                 header=False)
    y[te].to_csv(os.path.join(idp_dir, 'resp_te_index.txt'),
                 index=True,
                 sep=" ",
                 header=False)
    x[tr].to_csv(os.path.join(idp_dir, 'cov_tr.txt'),
                 index=False,
                 columns=x.columns[1:],
                 sep=" ",
                 header=False)
    x[te].to_csv(os.path.join(idp_dir, 'cov_te.txt'),
                 index=False,
                 columns=x.columns[1:],
                 sep=" ",
                 header=False)

    print(idp, lookup_UKB(field_ids=[idp])['Field'].to_string(), 'n =', len(y))
```

# Create train test basis
We are creating a B-spline basis for the train and test data. Different types of B-spline transformations can be considered. Many other options could also generate equivalent performance (e.g. fractional polynomials). The B-spline is sufficient to model the types of data that we expect to see in neuroimaging, (ie. relatively slow varying trends) and performs better than vanilla polynomials, which induce a global curvature. 

```python
data_dir = '/project_cephfs/3022017.02/projects/big_data/data/ukb_processed_4'

with open(os.path.join(data_dir, 'idp_ids.txt'), 'r') as f:
    for line in f:
        idp_ids.append(line.strip())
        
sys.path.append('/home/preclineu/chafra/Desktop/PCNtoolkit/')

xmin = 40 # boundaries for ages of UKB participants +/- 5
xmax = 85
B = create_bspline_basis(xmin, xmax)

# set mode 
suffices = ['tr','te']
```

## create dummy data for visualisation
This data is created to make the pretty normative modelling plots. 

```python
xx = np.arange(xmin,xmax,0.5)
X_dummy = np.zeros((len(xx), 3))
X_dummy[:,0] = 1  # male
X_dummy[:,2] = xx
np.savetxt(os.path.join(data_dir,'cov_male_dummy.txt'), X_dummy)
X_dummy = np.concatenate((X_dummy, np.ones((len(xx), 1))), axis=1)
np.savetxt(os.path.join(data_dir,'cov_int_male_dummy.txt'), X_dummy)
X_dummy = np.concatenate((X_dummy, np.array([B(i) for i in X_dummy[:,2]])), axis=1)
np.savetxt(os.path.join(data_dir,'cov_bspline_male_dummy.txt'), X_dummy)
```

## Loop over the IDPs

```python
for idp in idp_ids[:-1]:
    print('Configuring IDP:', idp)
    idp_dir = os.path.join(data_dir, idp)
    os.chdir(idp_dir)
    
    for suff in suffices:
        # load data matrices
        X = np.loadtxt(os.path.join(idp_dir, 'cov_'+ suff +'.txt')) 
        
        # create B-spline basis set (make sure the age column is selected!)
        Phi = np.array([B(i) for i in X[:,1]])
        
        # Save basis set with only B-spline
        Xz = np.concatenate((X, Phi), axis=1)
        np.savetxt(os.path.join(idp_dir, 'cov_bspline_z_' + suff + '.txt'), Xz)
        
        # add intercept
        X = np.concatenate((X, np.ones((X.shape[0],1))), axis=1) 
        np.savetxt(os.path.join(idp_dir, 'cov_int_' + suff + '.txt'), X)
    
        # basis set with bspline and intercept
        X = np.concatenate((X, Phi), axis=1)
        np.savetxt(os.path.join(idp_dir, 'cov_bspline_' + suff + '.txt'), X)
```

Make sure that your final covariate matrix is the shape you want it to be. In general the B-spline should be performed on the age variable (or any other continuous variable)
