- Principal Component Analysis for Extremes and Application to U.S. Precipitation Jiang et al.

```python
def calibration_descriptives(x):
  n = np.shape(x)[0]
  m1 = np.mean(x)
  m2 = sum((x-m1)**2)
  m3 = sum((x-m1)**3)
  m4 = sum((x-m1)**4)
  s1 = np.std(x)
  skew = n*m3/(n-1)/(n-2)/s1**3
  sdskew = np.sqrt( 6*n*(n-1) / ((n-2)*(n+1)*(n+3)) )
  kurtosis = (n*(n+1)*m4 - 3*m2**2*(n-1)) / ((n-1)*(n-2)*(n-3)*s1**4)
  sdkurtosis = np.sqrt( 4*(n**2-1) * sdskew**2 / ((n-3)*(n+5)) )
  semean = np.sqrt(np.var(x)/n)
  sesd = s1/np.sqrt(2*(n-1))
  cd = [skew, sdskew, kurtosis, sdkurtosis, semean, sesd]
  return cd
```

```python
blr_metrics = pd.DataFrame(columns = ['eid', 'NLL', 'EV', 'MSLL', 'BIC','Skew','Kurtosis'])
suffix = 'estimate'
for dirname in idp_paths[1:]:
    filename = main_path + dirname + '/blr_bspline_WarpSinArcsinh/'
    path_check = Path(os.path.join(filename+ '/Z_estimate.txt'))
    if path_check.is_file():
        NLL = np.loadtxt(os.path.join(filename + 'NLL_' + suffix + '.txt')).tolist()
        EV = np.loadtxt(os.path.join(filename + 'EXPV_' + suffix + '.txt')).tolist()
        MSLL = np.loadtxt(os.path.join(filename + 'MSLL_' + suffix + '.txt')).tolist()
        BIC = np.loadtxt(os.path.join(filename + 'BIC_' + suffix + '.txt')).tolist()
        Z = np.loadtxt(os.path.join(filename + 'Z_' + suffix + '.txt')).tolist()
        [skew, sdskew, kurtosis, sdkurtosis, semean, sesd] = calibration_descriptives(Z)
    blr_metrics.loc[len(blr_metrics)] = [dirname[:-4], NLL, EV, MSLL, BIC, skew, kurtosis]
```

### Dealing with NaN-values

There are several ways to deal with NaN values, we can take the average, median or mode and fill the value or drop the participants all together. Here we will take an in between method:
- drop participants that have over 90% missing data 
- then replace the remaining NaN values with a random value from the Gaussian distribution with mean zero and SD 1. 

In general imputing the data in this manner should limit the introduction of new patterns to the data, but of course other methods of imputation can be considered (for example using the mean or mode of the data).

```python
original_len = len(Z_df)
Z_df = Z_df[~(Z_df.isna().sum(axis=1)/np.shape(Z_df)[1]>0.9)] #remove subject with over 90% missing data
print(f'Number of subjects dropped {original_len-len(Z_df)}')
print(f'Total number of subjects {len(Z_df)}')
```

# Compute the TPDM @andre is this correct?
We want to summarize the dependence in the tails via the TPDM, a matrix of bivariate tail dependencies. For this framework the variables should be heavy-tailed with common tail index $\alpha=2$ (Jiang et al. 2020). To ensure this we:
1. Perform a rank transformation to obtain F (the estimated marginal cdf from the data)
2. Then we transform our F with G(x) the cdf of a Frechet random variable with scale 1 and $\alpha=2$ (Cooley and Thibaud 2018)

In the figures below an illustration of the marginal transformation is demonstrated.

To compare to the original code partly written by Emeric Thibaud see:
- MEVT_kmeans_pca.ipynb

## How to deal with positive and negative z-values ?
1. We can leave them as it is for now. (@andre this is what we do right now)
2. We could take the absolute of the Z_scores and look at the positive orthonant
3. We can look only at the positives and replace the negatives by zero and then repeat this analysis with the negative values, replacing the positives by zero. However, this makes the transformation look bad. 
We will start with option 1

See: Mapping the Heterogeneous Phenotype of Schizophrenia and Bipolar Disorder Using Normative Models Thomas Wolfers et al.

```python
%%R
#Z_scores[Z_scores > 0] <- 0 # Look only at negative or positive values (this leaves gaps in the matrix)
#Z_scores <- abs(Z_scores) # Take all values positive, maybe change later
```

```python
%%R
rank_trans <- 
function(x){
    I <- !is.na(x); y <- c(); 
    y[!I] <- NA; 
    y[I] <- rank(x[I])/(length(x[I])+1); return(y)
}
```

```python
%%R 
#Z_scores <- scale(Z_scores, center = TRUE, scale= FALSE)
Z_scores.rank<- apply(Z_scores,MARGIN = 2,rank_trans) # Margin=2 to perform over columns
Z_scores.rv2 <- sqrt(-1/log(Z_scores.rank)) 
```

# Eigenvalue decomposition TPDM 

@ andre, not entirely sure if I computed the contributions correctly, also the comparison the standard PCA is maybe redundant. We can discuss this. 

One of the goals of PCA is to understand the structure of observations and variables. We will create a TPDM PCA method.
Calculate the eigenvalues and vectors from the TPDM.
The first extreme PCs explains 17% of the scale.
In standard PCA, scale is the variance. The variance is the scale of the random variable from its center (mean). We look at the behavior in the tail and therefor will use the term scale to indicate the equivalent variance in the tail (Jiang et al. 2020). 


```python
%%R
eig <- eigen(SigmaHat)
PVE <- eig$values / sum(eig$values)
round(PVE, 2)
print(PVE[1]*100) # Explained scale or variance traditionally
eig$vectors[,1]  <- -eig$vectors[,1] # the first eigenvector is all negative so mirror it.
```

```python
%%R
get_eig<- function(eig_values){
    variance <- eig_values*100/sum(eig_values)
    cumvar <- cumsum(variance)
    eig_values <- data.frame(eigenvalue = eig_values, variance = variance, 
                      cumvariance = cumvar)
    colnames(eig_values) <- c("eigenvalue", "scale.percent", 
                     "cumulative.scale.percent")
  rownames(eig_values) <- paste0("Dim.", 1:nrow(eig_values))
  
  eig_values 
    }

fviz_eig<-function(eig, choice=c("variance", "eigenvalue"), geom=c("bar", "line"),
                  barfill="steelblue", barcolor="steelblue", linecolor = "black",
                  ncp=10, addlabels=FALSE, hjust = 0, 
                   main = NULL, xlab = NULL, ylab = NULL, 
                  ggtheme = theme_minimal(), ...){
    eig <- get_eig(eig)
    eig <-eig[1:min(ncp, nrow(eig)), , drop=FALSE]
    choice <- choice[1]
    if(choice=="eigenvalue") {
        eig <- eig[,1]
        text_labels <- round(eig,1)
        if(is.null(ylab)) ylab <- "Eigenvalue"
        }
    else if(choice=="variance") {
    eig <- eig[,2]
    text_labels <- paste0(round(eig,1), "%")
    }

  df.eig <- data.frame(dim = factor(1:length(eig)), eig=eig )
  
  extra_args <- list(...)
  bar_width <- extra_args$bar_width
  linetype <- extra_args$linetype
  if(is.null(linetype)) linetype <- "solid"
  
  p <- ggplot(df.eig, aes(dim, eig, group=1 ))
  if("bar" %in% geom) p <- p + geom_bar(stat="identity", fill= barfill, 
                                        color = barcolor, width = bar_width)
  if("line" %in% geom) p <- p + geom_line(color = linecolor, linetype = linetype)+
    geom_point(shape=19, color=linecolor)
  if(addlabels) p <- p + geom_text(label = text_labels, vjust=-0.4, hjust = hjust)
  
  if(is.null(main)) main <- "Scree plot"
  if(is.null(xlab)) xlab <- "Dimensions"
  if(is.null(ylab)) ylab <- "Percentage of explained variances"
  
  p <- p + labs(title = main, x = xlab, y = ylab)
  ggpubr::ggpar(p, ggtheme = ggtheme,  ...)
}
```

For more information on loadings and individual contributions to PCs see:
http://www.sthda.com/english/articles/31-principal-component-methods-in-r-practical-guide/112-pca-principal-component-analysis-essentials/#quantitative-variables
paper: Principal component analysis Herve Abdi and Lynne J. Williams


Sometimes loadings and eigenvectors are confused, but here loadings will be defined as:
loadings = eigenvector*sqrt(eigenvalues) (factor scores)

And the contributions are ctr=loadings^2/eigenvalue.

From there we can calculate the individual contributions of each variable to the PCs.

```python
%%R
visualize_contrib <- function(eig, main = NULL, PC_nr = 1, top=10){
    V<- eig$vectors
    coord.var <- t(t(as.matrix(V))*sqrt(eig$values)) # Get the loadings
    contrib.var <- t(t(coord.var^2)/eig$values) 
    contrib.var<-contrib.var[,1:2]
    rownames(contrib.var) = rownames(SigmaHat)
    contrib<-contrib.var[,PC_nr][order(-contrib.var[,PC_nr])][1:100]*100
    row_names_ordered <-rownames(contrib.var)[order(-contrib.var[,PC_nr])][1:100]
    df <- data.frame(name = factor(row_names_ordered, row_names_ordered), contrib 
        = contrib, stringsAsFactors = TRUE)
    bar_plot<-ggpubr::ggbarplot(df, x = "name", y = "contrib", fill="steelblue", color ="steelblue", sort.val ="desc", top = top, ylab ="Contributions (%)", main = main, xtickslab.rt = 90, ggtheme = theme_minimal())+ geom_hline(yintercept=100/length(eig_ss$values), linetype=2, color="red")

    return(list(data.frame(contrib[1:top]), bar_plot))
}   
```

An eigenvalue > 1 indicates that the PC accounts for more variance than one of the original variables.

```python
%%R
round_df <- function(x, digits) {
    # round all numeric variables
    # x: data frame 
    # digits: number of digits to round
    numeric_columns <- sapply(x, mode) == 'numeric'
    x[numeric_columns] <-  round(x[numeric_columns], digits)
    x
}
```

## Correlation circle
The plot above is also known as variable correlation plots. It shows the relationships between all variables. It can be interpreted as follow:
http://www.sthda.com/english/articles/31-principal-component-methods-in-r-practical-guide/112-pca-principal-component-analysis-essentials/
- Positively correlated variables are grouped together.
- Negatively correlated variables are positioned on opposite sides of the plot origin (opposed quadrants).
- The distance between variables and the origin measures the quality of the variables on the factor map. Variables that are away from the origin are well represented on the factor map.
